# calculating time taken to report a crime

```python
df_age
```

```python
df_age["date_occured_no_time"] = df_age['date_occurred'].dt.date
df_age["date_occured_no_time"]= pd.to_datetime(df_age["date_occured_no_time"])
```

```python
df_age["time_taken_reporting"] = (df_age['date_reported'] - df_age["date_occured_no_time"]).dt.days
```

```python
df_age["time_taken_reporting"].value_counts().head(10)
type(df_age["time_taken_reporting"])
```

```python
df_age.groupby('crime_description') ["time_taken_reporting"].mean().sort_values(ascending = True).head(10)
```
